```python
import cv2
import dlib
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = [15, 10]

# Load the pre-trained face detector
detector = dlib.get_frontal_face_detector()

# Load the shape predictor model
predictor = dlib.shape_predictor('/home/hask/Project/cv_homework/lesson9/data/shape_predictor_68_face_landmarks.dat')


img = cv2.imread('/home/hask/Project/cv_homework/lesson9/data/milan.jpeg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Detect faces in the image
faces = detector(img_rgb)

for face in faces:
    x, y, w, h = (face.left(), face.top(), face.width(), face.height())
    cv2.rectangle(img_rgb, (x, y), (x + w, y + h), (0, 255, 0), 2)

# Display the image with landmarks
plt.imshow(img_rgb)
plt.title('Face Landmarks')
plt.axis('off')
plt.show()
```
